Thanks for the suggestions! After looking at a few other responses today, it seems as though the logic of using a "counter" or "count" variable and subsequently creating the pivot table right after that step seemed to be a pretty obvious/common way one would re-frame this data. However today it occured to me that we can also implement the use of a pivot table later on, after we (a) create a new, separate dataframe that includes the total crime counts for each crime type and (b) another separate dataframe that contains "daily sums", and then merging the two. Creating these new dataframes also eliminates the necessity of creating the "counter" variable - we can achieve the same thing by utilizing the value_counts() function which is already available in the Pandas library. See below:

```python
# Install dfply module
!pip install dfply

import pandas as pd
import numpy as np
from dfply import *
import requests
dat = pd.read_csv("chicago_summer_2018_crime_data.csv") ##read in data from csv file
dat.head() # Preview the data

index = dat.index #create index variable
columns = dat.columns #create columns variable
values = dat.values #create values variable - we don't actually use these, I just wanted to create them for informational purposes

newframe = dat[['day', 'primary_type', 'month']]
newframe #re-frame / sub-select data

newframe.sort_values(by=['month'])  #just wanted to see what this would look like

crimes = newframe.primary_type.unique() ##also did this out of curiosity

dat.fillna(0)
newframe.fillna(0)

sums = newframe['primary_type'].value_counts() ##this is the important part of my response! i've created an entirely new dataframe called sums, which uses the value_counts() function to return total # times each crime was committed (cumulative)
sums = pd.Series.to_frame(sums, name="total") ##converts Series to a dataframe, creates header called "total"
sums['primary_type'] = sums.index ##creates 'primary_type' column, rather than only having it as the index - we'll need this when we merge later!
sums


```

```python
days = newframe['day'].unique() ##store all unique values (31 days)
len(days)

months = newframe['month'].unique() ##store months in months variable
len(months)

daily_sums = pd.DataFrame(dat.groupby(['day', 'primary_type'])['primary_type'].count()) #create dataframe called daily_sums that contains the totals for each crime for each day
daily_sums
```

```python
daily_sums.rename(columns = {'primary_type':'crimes_per_day'}, inplace = True)##rename column because it has the same column name as the Key that we will need to merge with later
pivot = daily_sums.pivot_table('crimes_per_day', index=['primary_type'],columns=['day']) ##NOW use a pivot table to change spread
pivot.fillna(0) 

merged = pd.merge(pivot, sums, on='primary_type') ##now we can merge our new "pivoted" daily sums table with the totals from "sums" based on the "primary_type" key
merged.fillna(0)

month_length = len(days) ##as Adam said, might not be best to hard-code with a number so I created this variable
month_length

for i in range(1,month_length):
    merged.iloc[:,i] = merged.iloc[:,i]/merged.loc[:,"total"] ##fill dataframe with proportions the same way I did - I still think this is an efficient way to calculate proportions.
    
merged
merged.fillna(0)

final = round(merged,2)
final.fillna(0)

final = final.drop(columns = ['total'])

final.fillna(0)

```
