# Regression

```python
final.min() ##find where there are zeros
```

```python
final.non_violence = final.non_violence + 1 ##add 1 to zero vars
```

```python
final.violence = final.violence + 1 ##add 1 to zero vars
```

% GDP Growth = Non-violence + violence + labor force + population

```python
final['non_violence'] = np.log(final['non_violence'])
final['violence'] = np.log(final['violence'])
final['Labor force, total'] = np.log(final['Labor force, total'])
final['Population, total'] = np.log(final['Population, total'])
##log everything to input into model
```

```python
X = final[['Labor force, total', 'Population, total', 'non_violence', 'violence']]
Y = final['GDP growth (annual %)']

X = sm.add_constant(X)

OLS = sm.OLS(Y, X).fit()

pred = OLS.predict(X)

##regression
```

```python
OLS.summary()
```

The T-statistic for the violence (logged) variable is high and p-value is low, so it seems to be statistically significant as a predictor - politically instability seems to affect GDP growth (with violence having a larger effect). I wonder if including an interaction variable for violence and population would alter the results - e.g. countries with higher populations might be prone more to violence
